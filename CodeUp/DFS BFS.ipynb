```python
# 전광판 전구 조작

graph_on = []
graph_off = []

n, m = 5, 6

'''
n, m = map(int, input().split())
import copy
for i in range(n):
    graph_on.append(list(map(int, input().split())))
graph_off = copy.deepcopy(graph_on)
'''

graph_on.append([0, 0, 1, 0, 1, 1])
graph_on.append([0, 1, 1, 0, 0, 0])
graph_on.append([0, 0, 1, 0, 0, 0])
graph_on.append([1, 1, 1, 1, 1, 1])
graph_on.append([0, 0, 0, 1, 0, 0])

graph_off.append([0, 0, 1, 0, 1, 1])
graph_off.append([0, 1, 1, 0, 0, 0])
graph_off.append([0, 0, 1, 0, 0, 0])
graph_off.append([1, 1, 1, 1, 1, 1])
graph_off.append([0, 0, 0, 1, 0, 0])

def dfs(graph, x, y, a, b):
    
    if (x <= -1 or x >= n or y <= -1 or y >= m):
        return False
    
    if graph[x][y] == a:
        graph[x][y] = b
        
        dfs(graph, x-1, y, a, b)
        dfs(graph, x, y-1, a, b)
        dfs(graph, x+1, y, a, b)
        dfs(graph, x, y+1, a, b)
        return True
    return False

on, off = 0, 0
for i in range(n):
    for j in range(m):
        if dfs(graph_on, i, j, 0, 1) == True:            
            on += 1
        if dfs(graph_off, i, j, 1, 0) == True:            
            off += 1            

print(on, off)
```

```python
# 전광판 전구 조작

n, m = 5, 6

graph_on = []
graph_off = []

'''
n, m = map(int, input().split())
import copy
for i in range(n):
    graph_on.append(list(map(int, input().split())))
graph_off = copy.deepcopy(graph_on)
'''

graph_on.append([0, 0, 1, 0, 1, 1])
graph_on.append([0, 1, 1, 0, 0, 0])
graph_on.append([0, 0, 1, 0, 0, 0])
graph_on.append([1, 1, 1, 1, 1, 1])
graph_on.append([0, 0, 0, 1, 0, 0])

graph_off.append([0, 0, 1, 0, 1, 1])
graph_off.append([0, 1, 1, 0, 0, 0])
graph_off.append([0, 0, 1, 0, 0, 0])
graph_off.append([1, 1, 1, 1, 1, 1])
graph_off.append([0, 0, 0, 1, 0, 0])

dx = [1,-1,0,0]
dy = [0,0,1,-1]

def dfs(graph, x, y, a, b):
    graph[x][y] = b
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]

        if 0 <= nx < n and 0 <= ny < m:
            if graph[nx][ny] == a:
                dfs(graph, nx, ny, a, b)
                
on, off = 0, 0
for i in range(n):
    for j in range(m):
        if graph_on[i][j] == 0:            
            dfs(graph_on, i, j, 0, 1)
            on += 1
            
        if graph_off[i][j] == 1:       
            dfs(graph_off, i, j, 1, 0)
            off += 1               

print(on, off)
```
